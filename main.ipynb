```python
import pandas as pd
import requests
import time
from io import StringIO
from tqdm import tqdm
```

```python
# Isso aqui Ã© para ler a lista gigantesca de url que o katana pegou
with open("links_apple_katana.txt") as arquivo:
    lista_url_bruto = arquivo.read().splitlines()

lista_url = [url for url in lista_url_bruto if url.count("-") >= 1 and url.count("/") == 7]
print(lista_url)
```

```python
lista_dfs = []
headers = {'User-Agent': 'Mozilla/5.0'}

for url in tqdm(lista_url, desc="Baixando e Organizando"):
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code != 200:
            continue
            
        tabelas = pd.read_html(StringIO(response.text))
        
        if len(tabelas) > 0:
            df_bruto = tabelas[0]
            
            # Transforma a tabela vertical em horizontal
            df_transposed = df_bruto.set_index(0).T.reset_index(drop=True)
            lista_dfs.append(df_transposed)
        
    except ValueError:
        pass 
    except Exception as e:
        tqdm.write(f"Erro em {url}: {e}")

    time.sleep(1)

if lista_dfs:
    iphone_df = pd.concat(lista_dfs, ignore_index=True)
    iphone_df.to_csv('iphone.csv', index=False)
    print("\nConcluÃ­do! Tabela limpa salva.")
else:
    print("\nNenhum dado capturado.")
```
